```python
tag_dict={}
for i in range(1,tagscsv_nparray.shape[0]):
  tag_dict[tagscsv_nparray[i][2]]=tag_dict.get(tagscsv_nparray[i][2],{})
  tag_dict[tagscsv_nparray[i][2]][tagscsv_nparray[i][1]]=tag_dict[tagscsv_nparray[i][2]].get(tagscsv_nparray[i][1],0)+1
tag_list=list(tag_dict)
tag_num=len(tag_list)
print(tag_dict)
print(tag_list)
tag_num
```

```python
count_list_2d_tagnum_movienum=[]
for i in range(tag_num):
  t=[]
  for j in range(movie_num):
    t_dict=tag_dict.get(tag_list[i],{})
    n=t_dict.get(movie_list[j],0)
    t.append(n)
  count_list_2d_tagnum_movienum.append(t)
count_nparray_2d_tagnum_movienum=np.array(count_list_2d_tagnum_movienum)
print(count_nparray_2d_tagnum_movienum)
print(count_nparray_2d_tagnum_movienum.shape)
print(np.argmin(np.amax(count_nparray_2d_tagnum_movienum,0)))
print(np.argmin(np.amax(count_nparray_2d_tagnum_movienum,1)))
```

```python
applied_list_2d_tagnum_movienum=[[int(j>0) for j in i ]for i in count_list_2d_tagnum_movienum]
applied_nparray_2d_tagnum_movienum=np.array(applied_list_2d_tagnum_movienum)
print(applied_nparray_2d_tagnum_movienum)
print(applied_nparray_2d_tagnum_movienum.shape)
```

```python
T,S,D_t = np.linalg.svd(applied_nparray_2d_tagnum_movienum, full_matrices=0)
print(T.shape,T)
print(S.shape,S)
print(D_t.shape,D_t)
```

```python
S_rate=[1]
for i in range(1,len(S)):
  if S[i-1]>0:
    S_rate.append(S[i]/S[i-1])
S_num=len(S)
i=1
while i<len(S_rate):
  if S_rate[i]/S_rate[i-1]<0.001:
    S_num=i
    break
  i=i+1
print(S_num,S[S_num-1],S[S_num])
```

```python
T_delete=np.delete(T,np.s_[S_num:],1)
print(T_delete.shape)
'''
S_delete=np.delete(S,np.s_[S_num:])
S1_delete=1/S_delete
S1_delete=np.diag(S1_delete)
S1_delete
print(S1_delete.shape)
'''
```

```python
dt_nparray_2d_t=np.identity(tag_num)
di_nparray_2d_t=np.transpose(applied_list_2d_tagnum_movienum)
dt1_nparray_2d_t=dt_nparray_2d_t.dot(T_delete) #.dot(S1_delete)
di1_nparray_2d_t=di_nparray_2d_t.dot(T_delete) #.dot(S1_delete)
print(dt1_nparray_2d_t.shape)
print(di1_nparray_2d_t.shape)
print(np.linalg.norm(dt1_nparray_2d_t[t]))
dt1_norm_list=[np.linalg.norm(dt1_nparray_2d_t[t]) for t in range(dt1_nparray_2d_t.shape[0])]
di1_norm_list=[np.linalg.norm(di1_nparray_2d_t[i]) for i in range(di1_nparray_2d_t.shape[0])]
print(len(dt1_norm_list),min(dt1_norm_list))
print(max(di1_nparray_2d_t[3]),di1_norm_list[3])
```

```python
a_is_zero=[0]
b_is_zero=[0]
def sim(a,b,a_is_zero,b_is_zero):
  if np.linalg.norm(a)==0:
    a_is_zero[0]=a_is_zero[0]+1
    return 0
  if np.linalg.norm(b)==0:
    b_is_zero[0]=b_is_zero[0]+1
    return 0
  return a.dot(b)/(np.linalg.norm(a)*np.linalg.norm(b))
lsi_sim_list_2d_tagnum_movienum=[[sim(dt1_nparray_2d_t[t],di1_nparray_2d_t[i],a_is_zero,b_is_zero) for i in range(movie_num)]for t in range(tag_num)]
lsi_sim_nparray_2d_tagnum_movienum=np.array(lsi_sim_list_2d_tagnum_movienum)
print(lsi_sim_nparray_2d_tagnum_movienum.shape)
print(a_is_zero[0],b_is_zero[0])
```
